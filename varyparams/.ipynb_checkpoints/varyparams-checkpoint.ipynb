```python
import pyccl as ccl
ccl.__version__
import numpy as np
import matplotlib as mpl
import matplotlib.pylab as plt
cmap = plt.cm.magma
```

```python
oc = np.linspace(.1, .9, 50)
ob = np.linspace(.03, .7, 50)
h = np.linspace(.55, .91, 50)
As = np.linspace(1.89e-9,2.5e-9,50)
ns = np.linspace(.87, 1.07, 50)
#ok = np.linspace() #not varying?
w0 = np.linspace(-3, -1, 50)
wa = np.linspace(-3, 1, 50)
```

```python
pz_path = '/home/s1/ggibbins/practice/lsst_like_dndz/' #where you saved the files from google drive

nside = 1024
ell = np.arange(0,  3*nside+1)
bias_z = [1.0, 1.1, 1.2, 1.5, 2.0]
```

```python
galtracers_dndz = []
for i in range(5):
    zz, nz = np.load(pz_path+'NZ_lensgalaxies_bin_'+str(i)+'.npy')
    galtracers_dndz.append(nz)

galtracers_dndz = np.array(galtracers_dndz)

wltracers_dndz = []
for i in range(5):
    zz, nz = np.load(pz_path+'NZ_sourcegalaxies_bin_'+str(i)+'.npy')
    wltracers_dndz.append(nz)

wltracers_dndz = np.array(wltracers_dndz)
```

```python
#Gal Vary oc

fig, axs = plt.subplots(5,sharex=True, figsize=(7,10)) #sharey=false to see differences in more detail (true shows scale better)

fig.suptitle("Galaxy PS Varying $\Omega_{c}$", fontsize=20, y = .94)

norm = mpl.colors.Normalize(vmin = oc[0], vmax = oc[49])


#Iterates through oc list to create a cross ps for each value
for z in range(5): #iterates thru z bins
    axs[z].set_ylabel(r'$C_{\ell}^{gg}$', fontsize=17)
    axs[z].set_title("Z-Bin "+str(z))
    for c in range(50): #iterates thru c cosmologies
        cosmo = ccl.Cosmology(Omega_c= oc[c], Omega_b=0.05, 
            h=0.7, A_s=2.1e-9, n_s= .96, Omega_k=0.0,
            w0 = -1, wa= 0, extra_parameters={"camb": {"dark_energy_model": "ppf"}}) #Creates 50 cosmologies varying w0 from -3 to 1
        gal = ccl.NumberCountsTracer(cosmo,False,(zz,galtracers_dndz[z]), bias= (zz,  bias_z[z]*np.ones(len(zz))))
        gal_auto = ccl.angular_cl(cosmo, gal, gal, ell)
        axs[z].plot(ell, gal_auto, c = cmap(norm(oc[c])))

plt.xscale('log')
plt.xlim(1,1500)
plt.xlabel(r'$\ell$', fontsize=20)

cb = fig.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap), ax = fig.get_axes())
cb.ax.tick_params(labelsize=11)
cb.set_label(label = '$\Omega_{c}$', size = 18)

plt.savefig('/home/s1/ggibbins/varyparams/fidplots/oc_gal')
```

```python
#SAVEDDDDD ON 6/20
#WL Vary oc

fig, axs = plt.subplots(5,sharex=True, figsize=(7,10)) #sharey=false to see differences in more detail (true shows scale better)

fig.suptitle("Weak Lensing PS Varying $\Omega_{c}$", fontsize=20, y = .94)

norm = mpl.colors.Normalize(vmin = oc[0], vmax = oc[49])

#Iterates through oc list to create a cross ps for each value
for z in range(5): #iterates thru z bins
    axs[z].set_ylabel(r'$C_{\ell}^{\kappa\kappa}$', fontsize=17)
    axs[z].set_title("Z-Bin "+str(z))
    for c in range(50): #iterates thru c cosmologies
        cosmo = ccl.Cosmology(Omega_c= oc[c], Omega_b=0.05, 
            h=0.7, A_s=2.1e-9, n_s= .96, Omega_k=0.0,
            w0 = -1, wa= 0, extra_parameters={"camb": {"dark_energy_model": "ppf"}}) #Creates 50 cosmologies varying w0 from -3 to 1
        lens = ccl.WeakLensingTracer(cosmo, (zz,wltracers_dndz[z]))
        lens_auto = ccl.angular_cl(cosmo, lens, lens, ell)
        axs[z].plot(ell, lens_auto, c = cmap(norm(oc[c])))

plt.xscale('log')
plt.xlim(1,1500)
plt.xlabel(r'$\ell$', fontsize=20)

cb = fig.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap), ax = fig.get_axes())
cb.ax.tick_params(labelsize=11)
cb.set_label(label = '$\Omega_{c}$', size = 18)

plt.savefig('/home/s1/ggibbins/varyparams/fidplots/oc_wl')
```
